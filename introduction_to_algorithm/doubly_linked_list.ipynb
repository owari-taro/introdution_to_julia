```julia
import Base:length
mutable struct Node
    prev::Union{Node,Nothing}
    key
    next::Union{Node,Nothing}
end



mutable struct LinkedList
    head::Node
end

function length(linked_list::LinkedList)
    current=linked_list.head
    count=0
    while current !=Nothing
        count+=1
        current=current.next
    end
    return count
end

function list_search(list::LinkedList,key)
    x=list.head
    while x !=nothing && x.key!=key
        x=x.next
    end
    return x
end

#insert new node at head of  linked list
function list_insert(linked_list::LinkedList,x::Node)
    x.next=linked_list.head
    if linked_list.head!=Nothing
        linked_list.head.prev=x
    end
    linked_list.haed=x
    x.prev=Nothing
end

    

function list_delete(linked_list::LinkedList,x::Node)
    if x.prev!=Nothing
        x.prev.next=x.next
    else
        linked_list.head=x.next
    end
    if x.next!=Nothing
        x.next.prev=x.prev
    end
end
```
