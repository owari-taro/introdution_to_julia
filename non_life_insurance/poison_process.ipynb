```julia
#using Distributions
using Plots
using Statistics
#probability function
poisson(λ,k)=exp(-λ)*(λ^k)/factorial(k)
λ=1
Δt=0.001
Δp=poisson(λ*Δt,1)
y=[0.0]
x=[0.0]
count=0
for  i in 1:10000
    push!(x,i*Δt)
    if rand()<Δp
        #print(i,":",count)
        count+=1
    end
    push!(y,count)
end

print(length(y))
plot(x,y,label="poisson-process λ=$λ Δt=$(Δt)")

```

```julia
using StatsPlots
#simulate first occurence time of poisson process
function simulate_occurence_time(λ;n=1,Δt=0.01)
    count=0
    Δp=poisson(λ*Δt,1)
    occurence_time=0
    for i in 1:n
        while true
            count+=1
            if rand()<Δp
                occurence_time+=count*Δt
                count=0
                break
            end
        end
    end
    return occurence_time
end

a=[simulate_occurence_time(2.0,n=10) for i in 1:10000]
@show mean(a)
@show var(a)
density(a,label = "Empirical Distribution")
```
